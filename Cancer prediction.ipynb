### we noticed a significant change in correlation between the two classes/

```python
def diff_in_corr (X1,X2):
    X=abs(X1-X2)
    print("the significant correlation diffrence")
    for i in range(X.shape[0]):
        for j in range(X.shape[0]):
            if(X.iloc[i,j]>=0.4):
                print("(",X.index[i],",",X.columns[j],") :" ,X.iloc[i,j])
```

```python
diff_in_corr(df_M.corr(),df_B.corr())
```

```python
df_M.shape
```

```python
def plot_diff_corr (X1,X2):
    X=abs(X1-X2)
    print("the significant correlation diffrence")
    for i in range(X.shape[0]):
        for j in range(X.shape[0]):
            if(X.iloc[i,j]>=0.4):
                plt.figure(figsize=(10,6))
                print("(",X.index[i],",",X.columns[j],") :" ,X.iloc[i,j])
                plt.scatter(df_M[df_M.columns[i]],df_M[df_M.columns[j]],c="RED")
                plt.scatter(df_B[df_M.columns[i]],df_B[df_M.columns[j]],c="BLUE")
                plt.xlabel(df_M.columns[i])
                plt.ylabel(df_M.columns[j])
                plt.show()

```

```python
plot_diff_corr (df_M.corr(),df_B.corr())
```

# Machine Learning algorithms

## without makin any preprocessing

```python
X=dfRaw.drop('diagnosis',axis=1)
y=np.array(dfRaw['diagnosis']).reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=2002)
X_train.shape,y_train.shape,X_test.shape,y_test.shape
```

```python
from sklearn.svm import SVC
clf=SVC()
clf.fit(X_train,y_train)
y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_pred,y_test),'\n',
accuracy_score(y_pred,y_test))
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler()

X_ros, y_ros = ros.fit_resample(X_train, y_train)
```

```python
from pandas.core.common import random_state
svc_ros=SVC(random_state=24)
svc_ros.fit(X_ros,y_ros)
y_pred_ros=svc_ros.predict(X_test)
```

```python
confusion_matrix(y_pred_ros,y_test)
```

```python
accuracy_score(y_pred_ros,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
DT=DecisionTreeClassifier(random_state=2002,criterion="entropy",min_samples_split=4,max_depth=4,)
DT.fit(X_train,y_train)
y_pred_DT=DT.predict(X_test)
```

```python
confusion_matrix(y_pred_DT,y_test)
```

```python
accuracy_score(y_pred_DT,y_test)
```
